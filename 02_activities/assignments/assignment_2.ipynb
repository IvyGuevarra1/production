# Preprocessing

Create two [Column Transformers](https://scikit-learn.org/stable/modules/generated/sklearn.compose.ColumnTransformer.html), called preproc1 and preproc2, with the following guidelines:

- Numerical variables

    * (Preproc 1 and 2) Scaling: use a scaling method of your choice (Standard, Robust, Min-Max). 
    * Preproc 2 only: 
        
        + Choose a transformation for any of your input variables (or several of them). Evaluate if this transformation is convenient.
        + The choice of scaler is up to you.

- Categorical variables: 
    
    * (Preproc 1 and 2) Apply [one-hot encoding](https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.OneHotEncoder.html) where appropriate.


+ The only difference between preproc1 and preproc2 is the non-linear transformation of the numerical variables.
    


### Preproc 1

Create preproc1 below.

+ Numeric: scaled variables, no other transforms.
+ Categorical: one-hot encoding.

```python
import pandas as pd
import numpy as np
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler, OneHotEncoder,QuantileTransformer

# separate numerical data and categorical to prepare for next step
numeric_fire = ['coord_x', 'coord_y', 'ffmc', 'dmc', 'dc', 'isi', 'temp', 'rh', 'wind', 'rain']
categorical_fire = ['month', 'day']

# do preproc1 as indicated 

preproc1 = ColumnTransformer(transformers=[
    ('num', StandardScaler(), numeric_fire),
    ('cat', OneHotEncoder(handle_unknown='ignore'), categorical_fire)
])
```

### Preproc 2

Create preproc1 below.

+ Numeric: scaled variables, non-linear transformation to one or more variables.
+ Categorical: one-hot encoding.

```python
#torn between using quantile and power transform (i saw you used this in the class) but i am curious of the quantile

preproc2 = ColumnTransformer(transformers=[
    ('num_non_linear', QuantileTransformer(output_distribution='normal'), numeric_fire),
    ('cat', OneHotEncoder(handle_unknown='ignore'), categorical_fire)
])
```

## Model Pipeline


Create a [model pipeline](https://scikit-learn.org/stable/modules/generated/sklearn.pipeline.Pipeline.html): 

+ Add a step labelled `preprocessing` and assign the Column Transformer from the previous section.

## Regressor

+ Use a regression model to perform a prediction. 

    - Choose a baseline regressor, tune it (if necessary) using grid search, and evaluate it using cross-validation.
    - Choose a more advance regressor, tune it (if necessary) using grid search, and evaluate it using cross-validation.
    - Both model choices are up to you, feel free to experiment.

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.ensemble import RandomForestRegressor

baseline_model = KNeighborsRegressor()
advanced_model = RandomForestRegressor(random_state=42)
```

```python
# Pipeline A = preproc1 + baseline
from sklearn.pipeline import Pipeline
pipe_a = Pipeline([('preprocessing', preproc1), ('regressor', baseline_model)])
```

```python
# Pipeline B = preproc2 + baseline
pipe_b = Pipeline([('preprocessing', preproc2), ('regressor', baseline_model)])
```

```python
# Pipeline C = preproc1 + advanced model
pipe_c = Pipeline([('preprocessing', preproc1), ('regressor', advanced_model)])
```

```python
# Pipeline D = preproc2 + advanced model
pipe_d = Pipeline([('preprocessing', preproc2), ('regressor', advanced_model)])
```

# Tune Hyperparams

+ Perform GridSearch on each of the four pipelines. 
+ Tune at least one hyperparameter per pipeline.
+ Experiment with at least four value combinations per pipeline.

```python
from sklearn.model_selection import GridSearchCV

# Baseline (KNN) I like this because we studied this
param_grid_baseline = {
    'regressor__n_neighbors': [3, 5, 11, 21],
    'regressor__weights': ['uniform', 'distance']
}
# Advanced (Random Forest) I have a bias on this because i worked with decision tree before but not RF
param_grid_advanced = {
    'regressor__n_estimators': [50, 100],
    'regressor__max_depth': [None, 10, 20],
    'regressor__min_samples_split': [2, 5]
}
```

```python
pipelines = {
    "Pipe A (Base/Scale)": (pipe_a, param_grid_baseline),
    "Pipe B (Base/NonLinear)": (pipe_b, param_grid_baseline),
    "Pipe C (Adv/Scale)": (pipe_c, param_grid_advanced),
    "Pipe D (Adv/NonLinear)": (pipe_d, param_grid_advanced)
}
```

```python
best_models = {}
results = {}

for name, (pipe, grid) in pipelines.items():
    #use MAE (Mean Absolute Error) I read that more robust to the outliers for the fires area
    search = GridSearchCV(pipe, grid, cv=5, scoring='neg_mean_absolute_error', n_jobs=-1)
    search.fit(X, y)
    
    best_models[name] = search.best_estimator_
    results[name] = -search.best_score_
    print(f"{name} | Best Params: {search.best_params_} | MAE: {results[name]:.4f}")
```

# Evaluate

+ Which model has the best performance?

```python
best_pipe_name = min(results, key=results.get)
best_model = best_models[best_pipe_name]
print(f"\n Best Performing Model: {best_pipe_name}")
```

# Export

+ Save the best performing model to a pickle file.

```python
import pickle

with open('best_forestfire_model.pkl', 'wb') as f:
    pickle.dump(best_model, f)
print("Model saved to 'best_forestfire_model.pkl'")
```
